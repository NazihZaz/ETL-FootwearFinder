```python
import pandas as pd
import sqlalchemy as sql
from sqlalchemy import create_engine
```

# Extract CSVs into DataFrames

```python
SCAthletic_file = "./Output/ShopComAthletic.csv"
SCAthletic_df = pd.read_csv(SCAthletic_file)
SCAthletic_df.head()
```

# Transform SCAthletic_df

```python
#Clean dataframe to match sql schema
SCAthletic_dfClean = SCAthletic_df.drop(['Unnamed: 0', 'ProductID', 'Brand', 'Image'], axis = 1)
SCAthletic_dfClean = SCAthletic_dfClean.reset_index()
SCAthletic_dfClean = SCAthletic_dfClean.rename(columns = {'index':'ID'})
SCAthletic_dfClean = SCAthletic_dfClean[['ID', 'Name', 'Price', 'Gender', 'URL']]
SCAthletic_dfClean
```
