```python
# Import Dependecies
import pandas as pd
import sqlalchemy as sql
from sqlalchemy import create_engine
from config import USERNAME, PASSWORD
```

# Extract CSVs into DataFrames

```python
SCAthletic_file = "./Output/ShopComAthletic.csv"
SCAthletic_df = pd.read_csv(SCAthletic_file)
SCAthletic_df.head()
```

```python
Nike_file = "./Output/nike.csv"
Nike_df = pd.read_csv(Nike_file)
Nike_df
```

# Transform DataFrames

```python
#Clean SCAthletic dataframe to match sql schema
SCAthletic_dfClean = SCAthletic_df[['Name', 'Price', 'Gender', 'URL']]
SCAthletic_dfClean
```

# Create database connection

```python
# Connext to the database
connection_string = f"{USERNAME}:{PASSWORD}@localhost:5433/FootwearFinder_db"
engine = create_engine(f'postgresql://{connection_string}')

# Inspect the columns of the shoes table
sql.inspect(engine).get_columns('shoes')
result = engine.execute("SELECT * FROM shoes")
sql.inspect(engine).get_columns('shoes')
```

# Load DataFrames into database

```python
# Load SCAthlectic_dfClean to the shoes table
SCAthletic_dfClean.to_sql(name='shoes', con=engine, if_exists='replace', index=False)
```

```python
# Load Nike_df to the shoes table
Nike_df.to_sql(name='shoes', con=engine, if_exists='replace', index=False)
```
